```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
csfont = {'fontname':'Arial'}
```

# DETECCION BOTELLA

```python
# Leer los datos del archivo
try:
    data = pd.read_csv('data_bottle.txt', sep=' ', header=None, names=['t','tstamp', 'mechanism', 'flag'])
    # Eliminar las filas con NaN
    data = data.dropna()
        # Mostrar la dimensión de la tabla
    print(f"La dimensión de la tabla es: {data.shape}")

    # Seleccionar un rango de datos
    #data = data.iloc[0:1000]

    # Mostrar la dimensión de la tabla
    print(f"La dimensión nueva de la tabla es: {data.shape}")

    # Verificar que los datos se leyeron correctamente
    print(data.tail())

    # Convertir los datos a numpy arrays si es necesario
    t = data['t'].values
    tstamp = data['tstamp'].values
    mechanism = data['mechanism'].astype(int).values
    flag = data['flag'].values
    
    # Crear el gráfico 
    plt.figure(figsize=(10, 5))
    plt.plot(t, mechanism)
    plt.title('Gráfico',**csfont)
    plt.xlabel('Tiempo',**csfont)
    plt.ylabel('bool',**csfont)
    plt.show()

    # Crear el gráfico
    plt.figure(figsize=(10, 5))
    plt.plot(t, flag)
    plt.title('Gráfico',**csfont)
    plt.xlabel('Tiempo',**csfont)
    plt.ylabel('bool',**csfont)
    plt.show()
    
except Exception as e:
    print(f"Error al leer el archivo: {e}")
    raise

```

# IMU

```python
# Leer los datos del archivo
try:
    data = pd.read_csv('data_imu.txt', sep=' ', header=None, names=['t','tstamp', 'ox', 'oy', 'oz','ow', 'ax', 'ay', 'az', 'lx', 'ly', 'lz'])
    # Eliminar las filas con NaN
    data = data.dropna()
        # Mostrar la dimensión de la tabla
    print(f"La dimensión de la tabla es: {data.shape}")

    # Seleccionar un rango de datos
    #data = data.iloc[0:1000]

    # Mostrar la dimensión de la tabla
    print(f"La dimensión nueva de la tabla es: {data.shape}")

    # Verificar que los datos se leyeron correctamente
    print(data.tail())

    # Convertir los datos a numpy arrays si es necesario
    t = data['t'].values
    tstamp = data['tstamp'].values
    ox = data['ox'].values
    oy = data['oy'].values
    oz = data['oz'].values
    ow = data['ow'].values
    ax = data['ax'].values
    ay = data['ay'].values
    az = data['az'].values
    lx = data['lx'].values
    ly = data['ly'].values
    lz = data['lz'].values

    # Crear el gráfico de orientación(ox,oy,oz,ow), 
    # velocidad angular(ax,ay,az) y aceleracion linal(lx,ly,lz)

    # Crear el gráfico de orientación
    plt.figure(figsize=(10, 5))
    plt.plot(t, ox, label='ox')
    plt.plot(t, oy, label='oy')
    plt.plot(t, oz, label='oz')
    plt.plot(t, ow, label='ow')
    plt.title('Gráfico de orientación',**csfont)
    plt.xlabel('Tiempo',**csfont)
    plt.ylabel('Orientación',**csfont)
    plt.legend(['ox', 'oy', 'oz', 'ow'])
    plt.show()

    # Crear el gráfico de velocidad angular
    plt.figure(figsize=(10, 5))
    plt.plot(t, ax, label='ax')
    plt.plot(t, ay, label='ay')
    plt.plot(t, az, label='az')
    plt.title('Gráfico de velocidad angular',**csfont)
    plt.xlabel('Tiempo',**csfont)
    plt.ylabel('Velocidad angular',**csfont)
    plt.legend(['ax', 'ay', 'az'])
    plt.show()

    # Crear el gráfico de aceleración lineal
    plt.figure(figsize=(10, 5))
    plt.plot(t, lx, label='lx')
    plt.plot(t, ly, label='ly')
    plt.plot(t, lz, label='lz')


    plt.title('Gráfico de aceleración lineal',**csfont)
    plt.xlabel('Tiempo',**csfont)
    plt.ylabel('Aceleración lineal',**csfont)
    plt.legend(['lx', 'ly', 'lz'])
    plt.show()
    
except Exception as e:
    print(f"Error al leer el archivo: {e}")
    raise

```

```python
# Seleccionar un rango de datos
data = data.iloc[0:1000]

# Mostrar la dimensión de la tabla
print(f"La dimensión nueva de la tabla es: {data.shape}")

# Verificar que los datos se leyeron correctamente
print(data.tail())

# Convertir los datos a numpy arrays si es necesario
t = data['t'].values
tstamp = data['tstamp'].values
ox = data['ox'].values
oy = data['oy'].values
oz = data['oz'].values
ow = data['ow'].values
ax = data['ax'].values
ay = data['ay'].values
az = data['az'].values
lx = data['lx'].values
ly = data['ly'].values
lz = data['lz'].values

# Crear el gráfico de orientación(ox,oy,oz,ow), 
# velocidad angular(ax,ay,az) y aceleracion linal(lx,ly,lz)

# Crear el gráfico de orientación
plt.figure(figsize=(10, 5))
plt.plot(t, ox, label='ox')
plt.plot(t, oy, label='oy')
plt.plot(t, oz, label='oz')
plt.plot(t, ow, label='ow')
plt.title('Gráfico de orientación',**csfont)
plt.xlabel('Tiempo',**csfont)
plt.ylabel('Orientación',**csfont)
plt.legend(['ox', 'oy', 'oz', 'ow'])
plt.show()

# Crear el gráfico de velocidad angular
plt.figure(figsize=(10, 5))
plt.plot(t, ax, label='ax')
plt.plot(t, ay, label='ay')
plt.plot(t, az, label='az')
plt.title('Gráfico de velocidad angular',**csfont)
plt.xlabel('Tiempo',**csfont)
plt.ylabel('Velocidad angular',**csfont)
plt.legend(['ax', 'ay', 'az'])
plt.show()

# Crear el gráfico de aceleración lineal
plt.figure(figsize=(10, 5))
plt.plot(t, lx, label='lx')
plt.plot(t, ly, label='ly')
plt.plot(t, lz, label='lz')


plt.title('Gráfico de aceleración lineal',**csfont)
plt.xlabel('Tiempo',**csfont)
plt.ylabel('Aceleración lineal',**csfont)
plt.legend(['lx', 'ly', 'lz'])
plt.show()
```

## MECANISMO

```python
# Leer los datos del archivo
try:
    data = pd.read_csv('data_mechanism.txt', sep=' ', header=None, names=['t', 'speed','sp'])
except Exception as e:
    print(f"Error al leer el archivo: {e}")
    raise

# Mostrar la dimensión de la tabla
print(f"La dimensión de la tabla es: {data.shape}")

# Seleccionar un rango de datos
#data = data.iloc[0:1000]

# Mostrar la dimensión de la tabla
print(f"La dimensión nueva de la tabla es: {data.shape}")

# Verificar que los datos se leyeron correctamente
print(data.tail())

# Convertir los datos a numpy arrays si es necesario
t = data['t'].values
speed = data['speed'].values
sp = data['sp'].values

# Crear el gráfico
plt.figure(figsize=(10, 5))

# Graficar 't' en el eje x y 'speed' en el eje y
plt.plot(t, speed, label='Mechanism', linestyle='-',linewidth=1.5)

# Graficar 't' en el eje x y 'sp' en el eje y
plt.plot(t, sp, label='referencia', linestyle='--',linewidth=3.0)

# Añadir título y etiquetas
plt.title('Gráfico de la velocidad de la cuenta del motor del mecanismo',**csfont)
plt.xlabel('t (seg)',**csfont)
plt.ylabel('Valores (cuentas/seg)',**csfont)
plt.legend()


# Mostrar el gráfico
plt.grid(True)
plt.show()
```
