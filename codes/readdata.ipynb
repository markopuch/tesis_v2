```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
csfont = {'fontname':'Arial'}
```

```python
# Leer los datos del archivo
try:
    data = pd.read_csv('datar.txt', sep=' ', header=None, names=['t', 'speed1', 'speed2', 'vrf_ticks', 'vrb_ticks'])
except Exception as e:
    print(f"Error al leer el archivo: {e}")
    raise

# Verificar que los datos se leyeron correctamente
print(data.head())

# Convertir los datos a numpy arrays si es necesario
t = data['t'].values
speed1 = data['speed1'].values
vrf_ticks = data['vrf_ticks'].values

# Crear el gráfico
plt.figure(figsize=(10, 5))

# Graficar 't' en el eje x y 'speed1' en el eje y
plt.plot(t, speed1, label='Rueda derecha delantera', linestyle='-',linewidth=1.5)

# Graficar 't' en el eje x y 'vrf_ticks' en el eje y
plt.plot(t, vrf_ticks, label='referencia', linestyle='--',linewidth=3.0)

# Añadir título y etiquetas
plt.title('Gráfico de la velocidad de la cuenta del motor derecha delantera',**csfont)
plt.xlabel('t (seg)',**csfont)
plt.ylabel('Valores (cuentas/seg)',**csfont)
plt.legend()


# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
speed2 = data['speed2'].values
vrb_ticks = data['vrb_ticks'].values
# Crear el gráfico
plt.figure(figsize=(10, 5))

# Graficar 't' en el eje x y 'speed1' en el eje y
plt.plot(t, speed2, label='Rueda derecha trasera', linestyle='-',linewidth=1.5)

# Graficar 't' en el eje x y 'vrf_ticks' en el eje y
plt.plot(t, vrb_ticks, label='referencia', linestyle='--',linewidth=3.0)

# Añadir título y etiquetas
plt.title('Gráfico de la velocidad de la cuenta del motor derecha trasera',**csfont)
plt.xlabel('t (seg)',**csfont)
plt.ylabel('Valores (cuentas/seg)',**csfont)
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
